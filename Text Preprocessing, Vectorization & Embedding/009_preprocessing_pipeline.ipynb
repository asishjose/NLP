```python
import re
import nltk
import spacy
import contractions

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
text = """
I can't believe this product is AMAZING!!! 😍 I've used it 3 times,
and it doesn’t disappoint. The build-quality is top-notch,
but it's kinda expensive... Would I buy it again? Yes, absolutely!!
"""
```

```python
text = text.lower()
```

```python
text = contractions.fix(text)
text
```

```python
# re.sub(r'\W+', '', text)
text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
text
```

```python
stop_words = set(stopwords.words('english'))
```

```python
tokens = word_tokenize(text)
tokens
```

```python
filtered_tokens = [word for word in tokens if word not in stop_words]
filtered_tokens
```

```python
stemmer = PorterStemmer()

stemmed_tokens = [stemmer.stem(word) for word in filtered_tokens]
stemmed_tokens
```

```python
def spacy_lemmatization(text):
    doc = nlp(text)
    lemmatized = [
        token.lemma_ for token in doc
        if not token.is_stop and not token.is_punct
    ]
    return lemmatized

lemmatized_tokens = spacy_lemmatization(text)
lemmatized_tokens
```

```python
final_text = " ".join(lemmatized_tokens)
final_text
```

```python
documents = [
    "cannot believe product amazing use 3 time not disappoint",
    "build quality top notch expensive",
    "product expensive but quality good",
    "absolutely amazing buy again"
]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count_vectorizer = CountVectorizer()
x_bow = count_vectorizer.fit_transform(documents)

count_vectorizer.vocabulary_
```

```python
x_bow.toarray()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()
X_tfidf = tfidf_vectorizer.fit_transform(documents)

tfidf_vectorizer.vocabulary_
```

```python
X_tfidf.toarray()
```

```python
from gensim.models import Word2Vec

model = Word2Vec(
    sentences=documents,
    vector_size=100,
    window=5,
    min_count=1,
    workers=4
)
```

```python
vector = model.wv["product"]
print(vector.shape)  # (100,)
print(vector[:10])   # first 10 dimensions
```

```python
model.wv.most_similar("amazing")
```

```python
import numpy as np

def document_vector(doc):
    return np.mean([model.wv[word] for word in doc], axis=0)

doc_vector = document_vector(documents[0])
print(doc_vector.shape)  # (100,)
```
