```python
from gensim.models import KeyedVectors
from gensim.downloader import load

glove_model = load('glove-wiki-gigaword-50')
word_pairs = [('learn', 'learning'), ('india', 'indian'), ('fame', 'famous')]

```

```python
# Compute similarity for each pair of words
for pair in word_pairs:
    similarity = glove_model.similarity(pair[0], pair[1])
    print(f"Similarity between '{pair[0]}' and '{pair[1]}' using GloVe: {similarity:.3f}")
```

```python
import numpy as np

def sentence_embedding(sentence, model):
    words = sentence.lower().split()
    vectors = [model[w] for w in words if w in model]
    
    if len(vectors) == 0:
        return np.zeros(model.vector_size)
    
    return np.mean(vectors, axis=0)

sentence = "learning is powerful"
embedding = sentence_embedding(sentence, glove_model)

print(embedding.shape)   # (50,)
```

```python
sentences = [
    "india is growing fast",
    "learning deep learning is fun"
]

X = np.array([sentence_embedding(s, glove_model) for s in sentences])
```
