# Speech Recognition

```python
import speech_recognition as sr

recognizer = sr.Recognizer()
```

```python
with sr.Microphone() as source:
    print("Speak something...")

    audio = recognizer.listen(source)

try:
    text = recognizer.recognize_google(audio)

    print("You said:", text)

except sr.UnknownValueError:
    print("Could not Understand audio!, Try Again...")

except sr.RequestError:
    print("Could not request results from Google Speech Recognition service!")
```

### Speech-to-Text from an Audio File

```python
import speech_recognition as sr

recognizer = sr.Recognizer()

with sr.AudioFile("speech.wav") as source:
    audio = recognizer.record(source)

try:
    text = recognizer.recognize_google(audio)
    print("Transcription:", text)

except sr.UnknownValueError:
    print("Audio not understood")

except sr.RequestError:
    print("API request failed")
```

# Speech Synthesis

### Text-to-Speech using gTTS (Google Text-to-Speech)

```python
from gtts import gTTS

text = "Hello, Welcome to natural language processing."

tts = gTTS(text=text, lang="en")

tts.save("tts_output.mp3")
print("Audio file saved")
```

## Speech Recognition + Speech Synthesis Pipeline

```python
import speech_recognition as sr
from gtts import gTTS
import os

recognizer = sr.Recognizer()

with sr.Microphone() as source:
    print("Speak something...")
    audio = recognizer.listen(source)

try:
    text = recognizer.recognize_google(audio)
    print("You said:", text)

    tts = gTTS(text=text, lang="en")
    tts.save("response.mp3")

    os.system("ffplay -nodisp -autoexit response.mp3")

except sr.UnknownValueError:
    print("Could not understand audio")

except sr.RequestError:
    print("Service unavailable")
```
