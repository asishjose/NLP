# Document Similarity

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity


documents = ['Document one text here', 'Document two text goes here']

tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

similarity = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix)
print(similarity)
```

The output [[1. 0.51014902]] from the cosine similarity function represents a similarity matrix where 1 indicates perfect similarity of a document with itself, and 0.51014902 shows moderate similarity between two different documents. The closer the value is to 1, the more similar the documents are.

```python
documents = [
    "machine learning is powerful",
    "AI and machine learning are useful",
    "football is a popular sport"
]

# Convert documents to TF-IDF vectors
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)

# Compute cosine similarity
similarity_matrix = cosine_similarity(tfidf_matrix)

print(similarity_matrix)
```

### Using Gensim Word2Vec

```python
from gensim.models import Word2Vec
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

docs = [
    "machine learning is powerful",
    "AI and machine learning are powerful"
]

#tokenize
tokenized_docs = [doc.split() for doc in docs]
model = Word2Vec(sentences=tokenized_docs, vector_size=100, window=5, min_count=1)

def document_vector(doc):
    vectors = [model.wv[word] for word in doc if word in model.wv]
    return np.mean(vectors, axis=0)

doc1_vec = document_vector(tokenized_docs[0])
doc2_vec = document_vector(tokenized_docs[1])

similarity = cosine_similarity([doc1_vec], [doc2_vec])

print("similarity:", similarity[0][0])
```

### Using Sentence embeddings instead of Word

Instead of averaging words, models directly produce document vectors.

Example models:

Sentence-BERT

Universal Sentence Encoder

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

model = SentenceTransformer("all-MiniLM-L6-v2")

docs = [
    "machine learning is powerful",
    "AI and machine learning are useful"
]

embeddings = model.encode(docs)

similarity = cosine_similarity([embeddings[0]], [embeddings[1]])

print(similarity)
```

# Document Search Engine

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

documents = [
    "Machine learning is used in AI.",
    "Deep Learning improves computer vision",
    "Football is a popular sport and Messi is a Greates of all time in it.",
    "Natural Language processing uses machine Learning"
]
```

```python
vectorizer = TfidfVectorizer()
doc_vectors = vectorizer.fit_transform(documents)
```

```python
def search(query):
    
    #convert query to vector
    query_vec = vectorizer.transform([query])

    #compute similarity
    similarity = cosine_similarity(query_vec, doc_vectors)

    scores = similarity.flatten()
    ranked_docs = scores.argsort()[::-1]


    print("Seacrch Results:\n")
    for i in ranked_docs[:3]:
        print(documents[i], "| score:", scores[i])
```

```python
search("ai")
```
