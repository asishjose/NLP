## Preprocessing

```python
text = "Hey!!! I’ve been having headaches for 3 days :( any idea what it could be???"
```

```python
import pandas as pd
import numpy as np
import re
```

```python
#handling the case
text_l = text.lower()
```

```python
# Expand contractions
import contractions

expanded_text = contractions.fix(text_l)
expanded_text
```

```python
# removing special characters
clean_text = re.sub(r'[^a-zA-Z0-9\s]', '', expanded_text)
clean_text
```

```python
# removing extra spaces and trailing spcaces
clean_text = re.sub(r'\s+', ' ', clean_text).strip()
clean_text
```

```python
# word tokenization

word_tokens = clean_text.split()
print(word_tokens)
```

```python
# Stopword Removal

from nltk.corpus import stopwords
stopword = stopwords.words('english')

def remove_stopwords(text):
    for word in text:
        if word in stopword:
            text.remove(word)
    return text

stopwords_removed = remove_stopwords(word_tokens)
print(stopwords_removed)

# df['review'].apply(remove_stopwords)
```

```python
# spelling correction

from textblob import TextBlob
incorrect_text = 'heeadachee'
textBlob = TextBlob(incorrect_text)
print(textBlob.correct())
```

```python
# spacy
import spacy
nlp = spacy.load("en_core_web_sm")
tokens = nlp(clean_text)
tokens
```

```python
# stemming & lemmatization
from nltk.stem import PorterStemmer, WordNetLemmatizer

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

def stem_words(text):
    return ' '.join([stemmer.stem(word) for word in text.split()])

def lemmatize_words(text):
    return ' '.join([lemmatizer.lemmatize(word, pos='v') for word in text.split()])

st = "walk walks walking walked"

stemmed = stem_words(st)
print(stemmed)

lemmatized = lemmatize_words(st)
print(lemmatized)
```

## Vectorization

```python
# BOW Vectorization

s1 = 'hey headache day'
s2 = 'headache day idea'

tokens1 = s1.split()
tokens2 = s2.split()

vocab = []

for word in tokens1+tokens2:
    if word not in vocab:
        vocab.append(word)

print(vocab)
```

### Bert Embedding

# NLP Applications

## Topic Modelling

```python
docs = [
    "the doctor treats patients in hospital",
    "nurse helps patients and doctor",
    "football match was exciting and team won",
    "player scored goal in match",
]
```

```python
from gensim.utils import simple_preprocess
from gensim.parsing.preprocessing import STOPWORDS

texts = [
    [word for word in simple_preprocess(doc) if word not in STOPWORDS]
    for doc in docs
]

print(texts)
```

```python
from gensim.corpora import Dictionary

dictionary = Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]
```

```python
from gensim.models import LdaModel

lda = LdaModel(
    corpus=corpus,
    id2word=dictionary,
    num_topics=2,
    passes=10,
    random_state=42
)
```

```python
lda.print_topics()
```

```python
for doc in corpus:
    print(lda.get_document_topics(doc))
```

## Similarity

```python
from gensim.models import Word2Vec
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

tokenized_docs = [
    ['doctor', 'treats', 'patients'],
    ['physician', 'treats', 'patients'],
    ['nurse', 'helps', 'patients']
]

model = Word2Vec(sentences=tokenized_docs,
                 vector_size=50,
                 min_count=1)

def doc_vector(doc):
    vectors = [model.wv[w] for w in model.wv]
    return np.mean(vectors, axis=0)

doc_vectors = np.array([doc_vector(doc) for doc in tokenized_docs])

def search(query):
    query_tokens = query.split()
    query_vec = doc_vectors(query_tokens).reshape(1, -1)

    sims = cosine_similarity(query_vec, doc_vectors)[0]
    ranked = sims.argsort()[::-1]

    return [(i, sims[i] for i in ranked]
```
